```python
# Estatistica Parametrica : os dados estão em alguma distribuição. geralmente a distribuição normal
# Estatistica não Parametrica: os dados estão em outra distribuição(desconhecida)
# Se os dados são "normais", usamos estatistica parametrica, Caso contrario a não parametrica
```

```python
#Baixndo as bibliotecas a serem utilizadas
import pandas as pd
#Plotagem de graficos
import seaborn as sns
#Bibliotecas estatistica
from scipy import stats
from scipy.stats import skewnorm
#Variaveis que receberão 1000 entradas numericas
dados_normais = stats.norm.rvs(size = 1000)
# se acaso vc alterar a variavel (a = -10) voce tera outro tipo de pllotagem dados enviesados
dados_nao_normais = skewnorm.rvs(a = 10, size = 1000)
```

```python
# Plotando dados normais 1000 numeros
sns.distplot(dados_normais);
```

```python
# Plotando dados não normais 1000 numeros
sns.distplot(dados_nao_normais);
```

# Teste de Shapiro-wilk

```python
# Nesse teste utiliza-se o valor (p = value) para interpretar o teste estatistico
# p <= alpha: rejeita a hipotese, não é normal
# p> alpha: não rejeita a hipotese, é normal
```

```python
# Baixando as bibliotecas para executar as verificações
from scipy.stats import shapiro
```

```python
#Vamos passar dois parametros um vazio outro será o (p)
_, p = shapiro(dados_normais)
p
```

```python
# Vamos montar um teste para verificar se os dados são "normais" 
#Passamos a variavel com parametros de medida
alpha = 0.05
if p > alpha:
    print('Distribuição normal')
else:
   print('Distribuição não normal')
```

```python
_, p = shapiro(dados_nao_normais)
p
```

```python
alpha = 0.05
if p > alpha:
    print('Distribuição normal')
else:
   print('Distribuição não normal')
```
