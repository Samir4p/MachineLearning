```python
# Baixando as bibliotecas a sererm usadas
import pandas as pd
```

```python
# Abraindo arquivo csv e passando separador e encoding
dataset = pd.read_csv('eleitorado_municipio_2020.csv', sep=';',encoding='latin1' )

# Pedindo para ler os primeiros registros do arquivo
dataset.head().T
```

```python
# ANALISANDO AS CIDADES COM MAIS DEFICIENTES
dataset.nlargest(5,'QTD_ELEITORES_SEMBIOMETRIA')
```

```python
# VAMOS DESCOBRIR A QUANTIDADE DE ELEITORES POR SEXO

print('Eleitoras:', dataset['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores:', dataset['QTD_ELEITORES_MASCULINO'].sum())
print('Sexo ñ Informado:', dataset['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# CRIAR AS VARIÁVEIS PARA ARMAZENAR OS VALORES

tot_eleitores = dataset['QTD_ELEITORES'].sum()
tot_feminino = dataset['QTD_ELEITORES_MASCULINO'].sum()
tot_masculino = dataset['QTD_ELEITORES_MASCULINO'].sum()
tot_nao_informado = dataset['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
print('Mulheres:',((tot_feminino / tot_eleitores) * 100).round(2))
print('Homens:', ((tot_masculino / tot_eleitores) * 100).round(2))
print('Não informados:', ((tot_nao_informado / tot_eleitores) * 100).round(2))
```

```python
# Quantos municipios com mais homens que mulheres ?

dataset[dataset['QTD_ELEITORES_MASCULINO'] > dataset['QTD_ELEITORES_FEMININO']].count()
```

```python
# vAMOS CONSTRUIR OUTRA COLUNA PARA RELACIONAR A QTDADE HOMEN X MULHER DO DATASET

dataset['RELACAO_FM'] = dataset['QTD_ELEITORES_MASCULINO'] / dataset['QTD_ELEITORES_FEMININO']
```

```python
# Quais os municipios com maior relacao m/f

dataset.nlargest(5, 'RELACAO_FM').T
```

```python
dataset.nsmallest(5, 'RELACAO_FM').T
```

```python
df_brasil = dataset[dataset['CD_PAIS'] == 1].copy()
df_brasil.head(3).T
```

```python
df_brasil.nlargest(5, 'RELACAO_FM').T
```

```python
df_brasil.nsmallest(5, 'RELACAO_FM').T
```

```python
%matplotlib inline

# Plotar um grafico de distribuição da relação f/m
df_brasil['RELACAO_FM'].plot.hist()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(df_brasil['RELACAO_FM'], bins = 100, color= 'red', kde = False)

# Alterando titulo do grafico
plt.title('Relação Eleitoras/Eleitores', fontsize=18)

# Alterando o eixo X
plt.xlabel('Eleitoras/Eleitores', fontsize=14)

# Alterando eixo Y
plt.ylabel('Frequência', fontsize=14)

# Vamos traçar uma linha tracejada para mostrar o maior indice da relação
plt.axvline(1.0, color='black', linestyle='--')
```

```python
sns.swarmplot(data=df_brasil, x='NM_REGIAO', y='RELACAO_FM')
plt.axhline(1.0, color='black', linestyle='--')
```
