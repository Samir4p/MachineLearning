# Site Camara dos Deputados
LINK -> https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla

```python
#Baixando as bibliotecas
import pandas as pd
#Request serve para abrir e buscar dados internet (api)
import requests
```

```python
# Obter os dados da API da Camara dos Deputados
# Variável para URL
url = 'https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla'

#Obtendo os dados da API
resp = requests.get(url)

# Verificando a Resposta da API
resp.text
```

```python
# Converter a resposta da API para DataFrame somente com os dados
df = pd.DataFrame(resp.json()['dados'])

# Mostrando os dados
df.head(10)
```

```python
# Qual numero atual de partidos
df['sigla'].count()
```

# Raspagem de Dados via wikipedia 

#link -> https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA

```python
# Importando as bibliotecas
import pandas as pd
import requests as rs

# URL
url = 'https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA'

# Requesitar a pagina e guardar a resposta
resposta = requests.get(url)

# Exibir os dados
table = pd.read_html(resposta.text)
table
```

```python
# Vamos olhar a tabela com numero de pagantes nas finais
#Essa é a 12 rabela dos dados
df = table[11]
# as primeiras linhas 
df.head()
```

```python
# Transformar uma coluna de object em inteiro 1º metodo
# df['Público pagante'] = df['Público pagante'].astype(int), não funcionou pois tem um ponto flutuante retirando ponto float

# Utilizando o comando 'replace' retiramos o ponto flaot e convertemos para int
df['Público pagante'] = df['Público pagante'].str.replace('.' , '').astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Exibindo grafico de linha com marcação em cada ponto
sns.pointplot(x='Ano', y='Publico pagante', data=df)

# Ajustando as legendas
plt.xticks(rotation=65)
plt.show()
```
