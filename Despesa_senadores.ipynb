# Analise dos dados de reembolso dos Senadores -  2020

```python
#Importando as bibliotecas
import pandas as pd
```

```python
# Abrindo arquivo csv 
# dessa forma (df = pd.read_csv('2020.csv')) vai dar um erro pois precisa passar parametros de separadores csv
# Utilizando a função skipprows=1 para retirar a primeira linha do arquivo
# Passando argumento (decimal = ',') avisamos ao df que este é o saparador decimal 
df = pd.read_csv('2020.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')
```

```python
# Mostrando as primeiras linhas do dataFrame
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Qual valor do reembolso

df['VALOR_REEMBOLSADO'].sum()
```

```python
df['SENADOR'].value_counts()
```

```python
# Quanto cada senador solicitou de reembolso em 2020 ?
# GroupBy para agrupar os dados semelhantes

df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
# qUAIS SÃO OS CINCO MAIORES VALORES DE REEMBOLSO

df.nlargest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Os menores valores reembolsados

df.nsmallest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Quantidade de reembolso por tipo de despesas

df['TIPO_DESPESA'].value_counts()
```
